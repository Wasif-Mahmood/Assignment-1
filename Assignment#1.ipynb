```python
# TASK 1

import math  

num = 0  
primeList=[]

def isprime(val):
    if val == 2:  
        primeList.append(2)  
        return True  
    elif val < 2:  
        return False  
    else:  
        for i in primeList:  
            if i > math.sqrt(val):
                break
            elif val % i == 0:
                return False
        primeList.append(val)
        return True

def rotations(n):
    answer=[]
    rotation=n
    while not rotation in answer:
        answer.append(rotation)
        rotation = int(str(rotation)[1:] + str(rotation)[0])
    return answer

limit=1000000
s=2
count=0
for i in range(s,limit):
    numList = rotations(i)
    valid = True
    for j in numList:
        if not isprime(j):
            valid = False
    if valid:
        num += 1
        count+=1

    print(num)
```
